```python
import time 
def evaluate_model(name, model, X_train, X_test, Y_train, Y_test):
    result = {}

    # Measure training time
    start_train = time.time()
    model.fit(X_train, Y_train)
    end_train = time.time()
    training_time = end_train - start_train

    # Measure prediction time
    start_pred = time.time()
    Y_pred = model.predict(X_test)
    end_pred = time.time()
    prediction_time = end_pred - start_pred

    # Metrics
    acc = accuracy_score(Y_test, Y_pred)
    prec = precision_score(Y_test, Y_pred)
    rec = recall_score(Y_test, Y_pred)
    f1 = f1_score(Y_test, Y_pred)

    # Confusion Matrix for FPR
    cm = confusion_matrix(Y_test, Y_pred)
    tn, fp, fn, tp = cm.ravel()
    fpr = fp / (fp + tn + 1e-6)  # to avoid division by zero

    # Store all results
    result["Model"] = name
    result["Accuracy"] = acc
    result["Precision"] = prec
    result["Recall"] = rec
    result["F1 Score"] = f1
    result["False Positive Rate"] = fpr
    result["Training Time (s)"] = training_time
    result["Prediction Time (s)"] = prediction_time

    return result
```

```python
import pandas as pd
import numpy as np
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
models = [
    ("Random Forest", RandomForestClassifier(n_estimators=100, random_state=42)),
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("XGBoost", XGBClassifier(use_label_encoder=False, eval_metric='logloss', verbosity=0))
]

results = []

for name, model in models:
    res = evaluate_model(name, model, X_train_scaled, X_test_scaled, Y_train, Y_test)
    results.append(res)

df_results = pd.DataFrame(results)
```
